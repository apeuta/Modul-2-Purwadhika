```python
df = pd.DataFrame({
    "x" : np.arange(10),
    "y" : np.arange(10),
    "z" : [0,1,0,1,0,0,1,0,1,2]
})
sb.relplot(data= df, x= "x", y= "y",
          hue= "z")
```

```python
sb.relplot(data= df, x= "x", y= "y",
          hue= "z", style= "z")
```

```python
sb.relplot(data= df, x= "x", y= "y",
          hue= "z", style= "z", size= "z")
```

```python
sb.barplot(data= df, x= "x", y="y")
```

```python
df = pd.DataFrame({
    "x" : np.arange(10),
    "y" : np.arange(10),
    "z" : ["Pria", "Wanita"] * 5
})
sb.barplot(data= df, x= "x", y= "y",
          hue= "z")
```

```python
sb.relplot(data= df, x= "x", y= "y",
          hue = "z", style= "z")

sb.barplot(data= df, x= "x", y= "y")
plt.show()
```

__DATASET__

```python
sb.get_dataset_names()
```

```python
df = sb.load_dataset("flights")
df
```

```python
sb.barplot(data= df, x="year", y="passengers")
plt.show()
```

```python
df_tahun = df.pivot(index= "year", values= "passengers", columns= "month")
df_tahun
```

```python
df_tahun = df_tahun.sum()
df_tahun = df_tahun.to_frame()
df_tahun
```

```python
sb.barplot(data= df_tahun, x = df_tahun.index, y= 0)
plt.show()
```

```python
sb.heatmap(df_tahun)
```

```python
sb.heatmap(df_tahun, annot= True, fmt= "d")
```

```python
df_korelasi = df.corr()
sb.heatmap(df_korelasi, annot=True, cmap= "hot_r")
```
