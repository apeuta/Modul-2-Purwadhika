### PLOT 3D BAR

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
fig = plt.figure()
p = plt.subplot(111, projection= "3d")

x = np.arange(10)
y = np.arange(10)
z = np.zeros(10)

dx = np.ones(10)
dy = np.ones(10)
dz = np.arange(10)

p.bar3d(x, y, z, dx, dy, dz)

p.set_xlabel("Sumbu X")
p.set_ylabel("Sumbu Y")
p.set_zlabel("Sumbu Z")

plt.show()
```

```python
fig = plt.figure()
p = plt.subplot(111, projection= "3d", title= "Plot Bar 3D")

x = np.zeros(8)
y = np.arange(8)
z = np.zeros(8)

dx = np.ones(8)
dy = np.ones(8)
dz = np.arange(8)

p.bar3d(x, y, z, dx, dy, dz,
       color =["w", "r", "orange", "yellow", "g", "b", "m", "purple"])

p.set_xlabel("Sumbu X")
p.set_ylabel("Sumbu Y")
p.set_zlabel("Sumbu Z")

plt.show()
```
