```python
df_new = pd.read_csv(
"databaru.csv",
na_values = ["-", "n.a"])

df_new = df_new.fillna(method = "ffill", axis= "index")

df_new
```

```python
df_new["gaji"] = df_new["gaji"].astype("int64")
df_new["usia"] = df_new["usia"].astype("int64")
df_new
```

```python
df_new.dtypes
```

```python
df_new = df_new.interpolate().fillna({
    "nama":"ANONIM"
})
df_new
```
