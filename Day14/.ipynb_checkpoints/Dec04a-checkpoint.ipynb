```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_ptabc.csv", 
                 delimiter=";", 
                 header = 2,
                 engine = "python",
                 skipfooter = 2,
                 index_col = 0
                )
df
```

delimiter untuk memasukkan batasan yang digunakan oleh csv. </br>
header untuk memulai dari baris tertentu. </br>
header bisa juga digantikan dengan skiprows. </br>
skipfooter untuk menghilangkan beberapa baris dari bawah. </br>
index_col = 0 atau .set_index("No.") untuk menjadikan kolom No. menjadi index acuan.

```python
df["Gaji"].describe().loc["max"]
```

```python
df.Gaji.max()
```

```python
df.Gaji.min()
```

```python
df.Gaji.mean()
```

```python
df.Gaji.sum()
```

```python
df.Gaji.mode()
```

```python
df.Gaji.median()
```

Menampilkan nama dan usia dengan gaji minimal

```python
df[["Nama", "Usia"]][df["Gaji"] == df["Gaji"].min()]
```

Menampilkan baris dengan usia maksimal

```python
df[df["Usia"] == df["Usia"].max()]
```

Menampilkan daftar baris dengan kondisi tertentu

```python
df[df["Usia"] != 35]
```

```python
df[df["Usia"] >= 25]
```

```python
df[(df["Usia"] > 25) & (df["Kota"] == "Jakarta")]
```

```python
df[df["Usia"].between(25, 30)]
```

```python
df[(df["Usia"] > 25) | (df["Kota"] == "Jakarta")]
```

& = And , | = Or

### Pokemon

```python
poke[poke["Speed"] == poke["Speed"].max()]
```
