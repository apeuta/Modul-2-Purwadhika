```python
import pandas as pd
```

```python
bps = pd.read_excel("bps-file.xlsx", header=3, skipfooter=1)
bps.rename(columns={'Unnamed: 0':'Provinsi'}, inplace=True)
bps
```

```python
bpsNew = pd.DataFrame({
    "No": (list(range(1,35)) * 2),
    "Provinsi": (list(bps["Provinsi"]) * 2),
    "Kawasan": ((["Tepi Laut"] * 34) + (["Bukan Tepi Laut"] * 34)),
    "2003": (list(bps[2003]) + list(bps["2003.1"])),
    "2005": (list(bps[2005]) + list(bps["2005.1"])),
    "2008": (list(bps[2008]) + list(bps["2008.1"])),
    "2011": (list(bps[2011]) + list(bps["2011.1"])),
    "2014": (list(bps[2014]) + list(bps["2014.1"])),
    "2018": (list(bps[2018]) + list(bps["2018.1"]))
})
bpsNew = bpsNew.set_index(["No", "Provinsi", "Kawasan"]).unstack().swaplevel(0,1,1).sort_index(1)
bpsNew
```
