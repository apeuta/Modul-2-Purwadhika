Selesaikan persamaan dan gambarkan dalam matplotlib </br>
x  - 2y +  z =  6 </br>
3x +  y - 2z =  4 </br>
7x - 6y -  z = 10

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection 
```

__Mencari hasil x,y,z dengan NumPy__

```python
a = np.array([[1, -2, 1], [3, 1, -2], [7, -6, -1]])
b = np.array([6, 4, 10])
hasil = np.linalg.solve(a, b)
x,y,z = hasil
print(f"Nilai X adalah {round(x)} \nNilai Y adalah {round(y)} \ndan Nilai Z adalah {round(z)}")
```

__Plotting plane with matplotlib__

```python
plt.figure('Ploting Plane', figsize=(24,8))

#PLOT 1
plt.subplot(131)
ax1 = plt.subplot(131, projection='3d')

# x - 2y + z = 6
x1= np.array([6, 0, 0])
y1= np.array([0, -3, 0])
z1= np.array([0, 0, 6]) 
ax1.scatter(x1, y1, z1, s=110)

titik1 = [list(zip(x1, y1, z1))]
polygon1 = Poly3DCollection(titik1, alpha=.45, facecolor='r')
plt.gca().add_collection3d(polygon1)

plt.title('x - 2y + z = 6')
ax1.set_xlabel('Nilai X')
ax1.set_ylabel('Nilai Y')
ax1.set_zlabel('Nilai Z')

#PLOT 2
plt.subplot(132)
ax2 = plt.subplot(132,projection='3d')

# 3x + y - 2z = 4
x2=np.array([4/3, 0, 0])
y2=np.array([0, 4, 0])
z2=np.array([0, 0, -2]) 
ax2.scatter(x2,y2,z2, s=110)

titik2 = [list(zip(x2, y2, z2))]
polygon2 = Poly3DCollection(titik2, alpha=.45, facecolor='g')
plt.gca().add_collection3d(polygon2)

plt.title('3x + y - 2z = 4')
ax2.set_xlabel('Nilai X')
ax2.set_ylabel('Nilai Y')
ax2.set_zlabel('Nilai Z')

#PLOT 3
plt.subplot(133)
ax3=plt.subplot(133,projection='3d')

# 7x - 6y - z = 10
x3=np.array([10/7, 0, 0])
y3=np.array([0, -10/6, 0])
z3=np.array([0, 0, -10]) 
ax3.scatter(x3,y3,z3, s=110)

titik3 = [list(zip(x3, y3, z3))]
polygon3 = Poly3DCollection(titik3, alpha=.45, facecolor='b')
plt.gca().add_collection3d(polygon3)

plt.title('7x - 6y - z = 10')
ax3.set_xlabel('Nilai X')
ax3.set_ylabel('Nilai Y')
ax3.set_zlabel('Nilai Z')

plt.show()
```
