```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfEXCL = pd.read_csv("EXCL.JK.csv", parse_dates = ["Date"], index_col = ["Date"])
dfEXCL = dfEXCL.sort_index()
dfEXCL = dfEXCL.resample("D").sum()
dfEXCL.iloc[:, :5] = dfEXCL.iloc[:, :5].replace(0, np.NaN)
dfEXCL = dfEXCL.fillna(method = "ffill")
```

```python
dfFREN = pd.read_csv("FREN.JK.csv", parse_dates = ["Date"], index_col = ["Date"])
dfFREN = dfFREN.sort_index()
dfFREN = dfFREN.resample("D").sum()
dfFREN.iloc[:, :5] = dfFREN.iloc[:, :5].replace(0, np.NaN)
dfFREN = dfFREN.fillna(method = "ffill")
```

```python
dfISAT = pd.read_csv("ISAT.JK.csv", parse_dates = ["Date"], index_col = ["Date"])
dfISAT = dfISAT.sort_index()
dfISAT = dfISAT.resample("D").sum()
dfISAT.iloc[:, :5] = dfISAT.iloc[:, :5].replace(0, np.NaN)
dfISAT = dfISAT.fillna(method = "ffill")
```

```python
dfTLKM = pd.read_csv("TLKM.JK.csv", parse_dates = ["Date"], index_col = ["Date"])
dfTLKM = dfTLKM.sort_index()
dfTLKM = dfTLKM.resample("D").sum()
dfTLKM.iloc[:, :5] = dfTLKM.iloc[:, :5].replace(0, np.NaN)
dfTLKM = dfTLKM.fillna(method = "ffill")
```

```python
date = dfEXCL.index.to_list()
closeEXCL = list(dfEXCL["Close"])
closeFREN = list(dfFREN["Close"])
closeISAT = list(dfISAT["Close"])
closeTLKM = list(dfTLKM["Close"])
```

```python
plt.figure(figsize = (11,6), dpi = 90)
plt.title("Harga Historis Saham Provider Telco Indonesia", size=15)
plt.plot(date,closeEXCL, "g", label="PT XL Axiata Tbk")
plt.plot(date, closeFREN, "c", label="PT Smartfren Telecom Tbk")
plt.plot(date, closeISAT, "b", label="PT Indosat Tbk")
plt.plot(date, closeTLKM, "r", label="PT Telekomunikasi Indonesia Tbk")
plt.legend(loc="upper left", ncol=4, mode="expand", borderaxespad = 0.)
plt.xlabel("Tanggal", size=12)
plt.ylabel("Rupiah (IDR)", size=12)
plt.grid(True)
plt.show()
```

```python
date2 = dfEXCL["2019-03-31":"2019-05-01"].index.to_list()
closeEXCL2 = list(dfEXCL["2019-03-31":"2019-05-01"]["Close"])
closeFREN2 = list(dfFREN["2019-03-31":"2019-05-01"]["Close"])
closeISAT2 = list(dfISAT["2019-03-31":"2019-05-01"]["Close"])
closeTLKM2 = list(dfTLKM["2019-03-31":"2019-05-01"]["Close"])
```

```python
plt.figure(figsize = (15,6), dpi = 80)
plt.title("Harga Historis Saham Provider Telco Indonesia bulan April 2019", size=15)
plt.plot(date2, closeEXCL2, "g", label="PT XL Axiata Tbk")
plt.plot(date2, closeFREN2, "c", label="PT Smartfren Telecom Tbk")
plt.plot(date2, closeISAT2, "b", label="PT Indosat Tbk")
plt.plot(date2, closeTLKM2, "r", label="PT Telekomunikasi Indonesia Tbk")
plt.legend(loc="upper left", ncol=4, mode="expand", borderaxespad = 0.)
plt.xticks(date2, rotation=45)
plt.xlabel("Tanggal", size=12)
plt.ylabel("Rupiah (IDR)", size=12)
plt.grid(True)
plt.show()
```
