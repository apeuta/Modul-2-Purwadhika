```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
pdd = pd.read_excel("penduduk.xls", na_values="-", header=3, skipfooter= 3)
pdd.rename(columns = {"Unnamed: 0": "Provinsi"}, inplace=True)
pdd.fillna(0, inplace=True)
pdd
```

```python
x = list((pdd["Provinsi"]))
y = list(pdd.columns[1:])
z1971 = list(pdd[1971])
z1980 = list(pdd[1980])
z1990 = list(pdd[1990])
z1995 = list(pdd[1995])
z2000 = list(pdd[2000])
z2010 = list(pdd[2010])
```

```python
fig = plt.figure(figsize=(20,15))
p = plt.subplot(111, projection= "3d")

xpos1 = list(range(33))
ypos1 = [0] * 33
zpos1 = [0] * 33

dx1 = [1] * 33
dy1 = [1] * 33
dz1 = z1971

p.bar3d(xpos1, ypos1, zpos1, dx1, dy1, dz1)

ypos2 = [1] * 33
dz2 = z1980

p.bar3d(xpos1, ypos2, zpos1, dx1, dy1, dz2)

ypos3 = [2] * 33
dz3 = z1990

p.bar3d(xpos1, ypos3, zpos1, dx1, dy1, dz3)

ypos4 = [3] * 33
dz4 = z1995

p.bar3d(xpos1, ypos4, zpos1, dx1, dy1, dz4)

ypos5 = [4] * 33
dz5 = z2000

p.bar3d(xpos1, ypos5, zpos1, dx1, dy1, dz5)

ypos6 = [5] * 33
dz6 = z2010

p.bar3d(xpos1, ypos6, zpos1, dx1, dy1, dz6)

p.set_xlabel("Provinsi")
p.set_ylabel("Tahun")
p.set_zlabel("Penduduk")
p.set_xticks(np.arange(33))
p.set_xticklabels(x, rotation=90)
p.set_yticks(np.arange(6))
p.set_yticklabels([1971, 1980, 1990, 1995, 2000, 2010])

plt.show()
```

```python
x = list((pdd["Provinsi"]))
y = list(pdd.columns[1:])
listTahun = [1971, 1980, 1990, 1995, 2000, 2010]

fig = plt.figure(figsize=(20,15))
p = plt.subplot(111, projection= "3d")

x1 = list(range(33))
z1 = [0] * 33

dx = [1] * 33
dy = [1] * 33

for i in range(len(listTahun)):
    y = []
    y = [i] * len(x)
    for j in range(len(listTahun)):
        dz = []
        dz = list(pdd[listTahun[j]])
    p.bar3d(x1, y, z1, dx, dy, dz)

p.set_xlabel("Provinsi")
p.set_ylabel("Tahun")
p.set_zlabel("Penduduk")

p.set_xticks(np.arange(33))
p.set_xticklabels(x, rotation=90)
p.set_yticks(np.arange(6))
p.set_yticklabels(listTahun)

plt.show()
```
