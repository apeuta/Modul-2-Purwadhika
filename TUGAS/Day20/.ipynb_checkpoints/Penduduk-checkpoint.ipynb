```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
pdd = pd.read_excel("penduduk.xls", na_values = ['-'], header=3, skipfooter= 3)
pdd
```

```python
fig = plt.figure()
p = plt.subplot(111, projection= "3d")

x = np.arange(33)
y = ["1971"] * len(x)
z = np.zeros(len(x))

dx = np.ones(len(x))
dy = np.ones(len(x))
dz = pdd[1971]

p.bar3d(x, y, z, dx, dy, dz)

p.set_xlabel("Sumbu X")
p.set_ylabel("Sumbu Y")
p.set_zlabel("Sumbu Z")

plt.show()
```
