```python
import numpy as np
import pandas as pd
import folium
import requests
```

```python
pokemon = pd.read_csv("pokemon.csv")
poke = pokemon[:1000]
poke
```

```python
num = poke["num"].to_list()
name = poke["name"].to_list()
lat = poke["lat"].to_list()
lng = poke["lng"].to_list()

point = list(zip(lat,lng))
```

```python
popup = []
for i in num :
    poke_url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    getdata = requests.get(poke_url)
    output = getdata.json()
    img_src = output["sprites"]["front_default"]
    popup.append(img_src)
```

```python
sf_map = folium.Map(
    location = [37.5082,-122.2854],
    zoom_start = 10.25)

for i in range(len(num)):
    nama = name[i]
    latlong = point[i]
    img_link = folium.features.CustomIcon(popup[i], icon_size=(25,25))
    folium.Marker(
        location = latlong,
        tooltip = f"<b>{nama}</b>",
        icon = img_link
    ).add_to(sf_map)
sf_map
```
