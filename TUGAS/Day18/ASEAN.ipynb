```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
```

```python
dbku = mysql.connector.connect(
host = "localhost",
user = "root",
passwd = "12345",
database = "world")

query = "select Name negara_ASEAN, Population, GNP, SurfaceArea from country where Region = 'Southeast Asia' order by negara_ASEAN"
df = pd.read_sql(query, dbku)
df
```

```python
plt.figure(figsize= (14,8))
plt.bar(df["negara_ASEAN"], df["Population"], 
        color=["b","orange","g","r","purple","brown","pink","gray","yellow","cyan","magenta"])

plt.title("Populasi Negara ASEAN", fontdict={
    "family":"arial black", "size":20,
    "color":"k"
})

#menampilkan label
plt.xlabel("Negara", fontsize=15)
plt.ylabel("Populasi (x 100jt jiwa)", fontsize=15)

#menampilkan value dari bar
for i in df["Population"]:
    plt.text(df[df["Population"] == i]["negara_ASEAN"], i, i, ha="center", va="bottom")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.pie (df["Population"], labels=df["negara_ASEAN"], radius=2.5, 
        colors=["b","orange","g","r","purple","brown","pink","gray","yellow","cyan","magenta"],
        autopct = "%1.1f%%")
plt.title("Persentase Penduduk ASEAN", fontdict={
    "family":"arial black", "size":20,
    "color":"k"
}, pad=150)
plt.show()
```

```python
plt.figure(figsize= (14,8))
plt.bar(df["negara_ASEAN"], df["GNP"], 
        color=["b","orange","g","r","purple","brown","pink","gray","yellow","cyan","magenta"])

plt.title("Pendapatan Bruto Negara ASEAN", fontdict={
    "family":"arial black", "size":20,
    "color":"k"
})

#menampilkan label
plt.xlabel("Negara", fontsize=15)
plt.ylabel("GNP (USD)", fontsize=15)

#menampilkan value dari bar
for i in df["GNP"]:
    plt.text(df[df["GNP"] == i]["negara_ASEAN"], i, i,  ha="center", va="bottom")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.pie (df["SurfaceArea"], labels=df["negara_ASEAN"], radius=2.5, 
        colors=["b","orange","g","r","purple","brown","pink","gray","yellow","cyan","magenta"],
        autopct = "%1.1f%%")
plt.title("Persentase Luas Daratan ASEAN", fontdict={
    "family":"arial black", "size":20,
    "color":"k"
}, pad=150)
plt.show()
```
