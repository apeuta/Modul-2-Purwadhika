__X ticks & Y Ticks__

__DIAGRAM BATANG__

```python
df = pd.DataFrame({
    "x": [1,2,3,4,5],
    "y": [9,5,7,3,6],
    "z": [2,3,4,5,3]
})

#menampilkan dalam bentuk bar chart
plt.bar(df.x, df.y, color="pink", edgecolor="black", linewidth=5)

#menampilkan judul
plt.title("Ini Bar Chart", fontdict={
    "family":"arial black", "size":20,
    "color":"k"
})

#menampilkan label
plt.xlabel("Ini X", fontsize=15)
plt.ylabel("Ini Y", fontsize=15)

#menampilkan value dari bar
for i in df.y:
    plt.text(df.x[list(df.y).index(i)], i+.3, i)

# #Opsi 2
# for i in df.y:
#     plt.text(df[df["y"] == i]["x"], i+.3, i)

#menampilkan grid
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    "x": [1,2,3,4,5],
    "y": [9,5,7,3,6],
    "z": [2,3,4,5,3]
})
plt.barh(df.x, df.y, color=["g", "pink", "r", "yellow", "b"], edgecolor="black", linewidth=5)
plt.title("Ini Bar Chart", fontdict={
    "family":"arial black", "size":20,
    "color":"k"
})
plt.xlabel("Ini X", fontsize=15)
plt.ylabel("Ini Y", fontsize=15)
plt.grid(True)
plt.show()
```

__Posisi Bar depan belakang__

```python
df = pd.DataFrame({
    "x": [1,2,3,4,5],
    "y": [9,5,7,3,6],
    "z": [2,3,4,5,3]
})

#menampilkan dalam bentuk bar chart
plt.bar(df.x, df.y, color="pink", edgecolor="black", linewidth=5)
plt.bar(df.x, df.z, color="k", edgecolor="black", linewidth=5)

#menampilkan value dari bar
for i in df.y:
    plt.text(df.x[list(df.y).index(i)], i+.3, i, size = 15)
for i in df.z:
    plt.text(df.x[list(df.z).index(i)], i+.3, i, color="r", size=15)
plt.show()
```

__Posisi bar stack atas bawah__

```python
df = pd.DataFrame({
    "x": [1,2,3,4,5],
    "y": [9,5,7,3,6],
    "z": [2,3,4,5,3]
})

#menampilkan dalam bentuk bar chart
plt.bar(df.x, df.y, bottom = df.z, yerr=df.describe()["y"].loc["std"], color="pink", edgecolor="black", linewidth=5)
plt.bar(df.x, df.z, color="k", edgecolor="black", linewidth=5)

plt.show()
```

Nilai error suatu data bisa didefinisikan dengan yerr, </br>
yerr=df.describe()["y"].loc["std"]

</br>__SCATTER__

```python
#membuat marker
plt.plot([1,2,3],[4,5,6], "r*", markersize=10)

#membuat scatter
plt.scatter([1,2,3], [3,4,5], color="b")

plt.show()
```

```python
#membuat line
plt.plot([1,2,3],[4,5,6], "r", markersize=10)

#membuat scatter
plt.scatter([1,2,3], [4,5,6], color="b", s=100)

plt.show()
```

Defaultnya, Scatter berada di belakang line, harus menggunakan zorder.

```python
#membuat line
plt.plot([1,2,3],[4,5,6], "r", zorder=2)

#membuat bar
plt.bar([1,2,3],[4,5,6], color="yellow", zorder=1)

#membuat scatter
plt.scatter([1,2,3], [4,5,6], color="b", s=100, zorder=3)

plt.show()
```

__DIAGRAM PIE__

```python
persentase = [25,55, 20]
keterangan = ["Tim A", "Tim B", "Tim C"]

plt.pie (persentase, labels=keterangan,
        colors=["pink", "cyan", "yellow"], startangle=90,
        shadow= True, counterclock=False,
        textprops={"size":12, "color":"k"},
        autopct = "%1.1f%%")

# plt.show()
```

__Mengubah warna persentase, dengan menembak lokasi list__

```python
persentase = [25,55, 20]
keterangan = ["Tim A", "Tim B", "Tim C"]

#define jadi 3 list
x,y,z = plt.pie (persentase, labels=keterangan,
        colors=["pink", "cyan", "yellow"], startangle=90,
        shadow= True, counterclock=False,
        textprops={"size":12, "color":"k"},
        autopct = "%1.1f%%",
        explode = (0.2, 0.1, 0))

#loop color
for i in z:
    i.set_color("green")

plt.show()
```

__HISTOGRAM__

```python
usia = [22,21,32,42,32,45,65,55,43,22,21,34,43,44]
par = [15,20,25,30,40,45,50,55,60,65]

plt.hist(usia, par, color="pink")
plt.xticks(np.arange(15,66,5))
plt.show()
```
