### Matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [1,3,5,2,4,6,8,7,3]
z = np.array([2,4,6,8,1,3,5,7,9])
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("X vs Y")
plt.xlabel("Nilai X")
plt.ylabel("Nilai Y")
plt.grid(True)
plt.legend(["Garis XY"])
plt.show()
```

title = judul grafik </br>
xlabel = label di axis x </br>
ylabel = label di axis y </br>
grid = menambahkan grid di dalam grafik </br>
legend = menambahkan informasi dari sebuah garis

__Untuk menampilkan 2 garis__ </br>
Bisa dengan mendeklarasikan 2x

```python
plt.plot(x,y)
plt.plot(x,z)
plt.title("Testing 1")
plt.xlabel("Nilai X")
plt.ylabel("Nilai Y/Z")
plt.grid(True)
plt.legend(["Garis XY", "Garis XZ"])
plt.show()
```

Atau dideklarasikan beruntun

```python
plt.plot(x,y, x, z)
plt.title("Testing 2")
plt.xlabel("Nilai X")
plt.ylabel("Nilai Y/Z")
plt.grid(True)
plt.legend(["Garis XY", "Garis XZ"])
plt.show()
```

__Matplotlib dengan 3 input dan fungsi berkaitan__

```python
x = np.array(range(1,10))
y = x ** 2
z = x ** 3
plt.plot(x, x, x, y, x, z)
plt.title("Testing 3")
plt.xlabel("Nilai X")
plt.ylabel("Nilai X/Y/Z")
plt.grid(True)
plt.legend(["Linier", "Kuadrat", "Pangkat 3"])
plt.show()
```

__Editing warna dengan coding__ </br>
r = Red, g = Green, b = Blue, c = Cyan, m = Magenta, y = Yellow, k = Black

```python
x = np.array(range(1,10))
y = x ** 2
z = x ** 3
plt.plot(x, x, "k", x, y, "pink", x, z, "#FF5733")
plt.title("Testing 3")
plt.xlabel("Nilai X")
plt.ylabel("Nilai X/Y/Z")
plt.grid(True)
plt.legend(["Cyan", "Magenta", "Blue"])
plt.show()
```

Nama bisa dipanggil dari nama resmi, maupun Hex Code dari https://htmlcolorcodes.com

__Editing bentuk garis dan marker__ </br>
":" = titik-titik di sepanjang garis </br>
"-" = garis biasa </br>
"--" = garis patah-patah </br>
"-." = garis titik </br>
"s" = kotak di node</br>
"^" = segitiga di node</br>
"v" = segitiga terbalik</br>
"p" = pentagon</br>
"D" = diamond</br>
"h" = hexagon</br>
"x" = x</br>
"*" = bintang di node</br>
"." = titik kecil di node </br>
"o" = titik besar di node

```python
x = np.array(range(1,10))
y = x ** 2
z = x ** 3
plt.plot(x, x, "r-.^", x, y, "g--*", x, z, "b:o")
plt.title("Testing 3")
plt.xlabel("Nilai X")
plt.ylabel("Nilai X/Y/Z")
plt.grid(True)
plt.legend(["Cyan", "Magenta", "Blue"])
plt.show()
```

__Bentuk Baku__

```python
plt.plot(x,x, color="b", marker="*", linestyle= ":", markersize=15, markerfacecolor="r", markeredgecolor="k", linewidth=3, label="Data")
plt.grid(True, color="pink", linestyle="-.")
plt.legend()
plt.subplots_adjust(top=0.55, left=0.1)
plt.text(8,2, "ini titik 8,2")
plt.show()
```

__Memberi info pada tiap node dengan menggunakan for loop__

```python
plt.plot(x,x ** 2, "-*")
for i in x**2:
    plt.text(x[list(x**2).index(i)] + .1, i-.3, i)
plt.grid(True)
plt.show()
```

__Memberi annotation__

```python
plt.plot(x,x ** 3, "-*")
for i in x**3:
    plt.text(x[list(x**3).index(i)] + .1, i-.3, i)
plt.annotate(
    "Highest Value!", xy = (9,729), xytext=(6,600),
    arrowprops = {"facecolor" : "k", "shrink" : .1})
plt.show()
```
