### 3D PLOT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d
```

```python
fig = plt.figure()
p = plt.subplot(111, projection= "3d")
data = range(11)
x = np.array(data)

p.plot_wireframe(x, x, x.reshape(1,-1), color="r", linestyle=":")
p.scatter(x, x, x, marker= "*", color= "g", s=100)

plt.show()
```
